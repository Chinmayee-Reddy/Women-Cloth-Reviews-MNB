**Title of Project:**
Women Cloth Reviews prediction with Multi Nomial Naive Bayes

Objective:To build a text classification model that predicts the sentiment (positive, negative, or neutral) or specific categories (such as fit, style, quality, etc.) of women's clothing reviews using Multinomial Naive Bayes.

**Import Library**

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

**Import Dataset**

```python
df=pd.read_csv('https://raw.githubusercontent.com/YBIFoundation/MachineLearning/refs/heads/main/Dataset/Women%20Clothing%20E-Commerce%20Review.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

**Missing Values**

```python
df.isna().sum()
```

```python
df[df['Review']==""]=np.NaN
```

```python
df['Review'].fillna("No Review",inplace=True)
```

```python
df.isna().sum()
```

```python
df['Review']
```

**Define Target(y) and Feature(x)**

```python
df.columns
```

```python
x=df['Review']
```

```python
y=df['Rating']
```

```python
df['Rating'].value_counts()
```

**Train Test Split**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,stratify=y,random_state=2529)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

**Get Feature Text Conversion To Tokens**

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(lowercase=True,analyzer='word',ngram_range=(2,3),stop_words='english',max_features=5000)
```

```python
x_train=cv.fit_transform(x_train)
```

```python
cv.get_feature_names_out()
```

```python
x_train.toarray()
```

```python
x_test=cv.fit_transform(x_test)
```

```python
cv.get_feature_names_out()
```

```python
x_test.toarray()
```

**Get Model Trained**

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model=MultinomialNB()
```

```python
model.fit(x_train,y_train)
```

**Get Model Predicted**

```python
y_pred=model.predict(x_test)
```

```python
y_pred.shape
```

```python
y_pred
```

**Get Probability of each Predicted Class**

```python
model.predict_proba(x_test)
```

**Get Model Evaluated**

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
